```
month = 8
path = '/content/drive/My Drive/PremierLeague/1819/01Fixtures/pl1819_fixture_' + str(month) + '.html'
with open(path,encoding="utf-8") as f:
    soup = BeautifulSoup(f, 'html.parser')
soup.select_one('#tournament-fixture > tbody > tr:nth-child(1) > th').string
```

```
match_idlist = []
count = 0
for i in  range(1,100):
  try:
    match_url = soup.select_one('#tournament-fixture > tbody > tr:nth-child(' + str(i) + ') > td.result > a')
    match_id = re.findall('[0-9]+',str(match_url))
    match_idlist.append(int(match_id[1]))
    count = 0
  except:
    if count == 1:
      break
    else:
      count = 1
```
