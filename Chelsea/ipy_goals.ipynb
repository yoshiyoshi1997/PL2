```python
import pandas as pd
import numpy as np
from pandas import DataFrame
from pandas import Series
import matplotlib
import matplotlib.pyplot as plt
```

```python
goals = pd.read_csv('goals.csv')
goals = goals.rename(columns={'Unnamed: 0':'homeway'})
goals
```

```python
goals[goals.team == 'Chelsea']
```

```python
goals[goals.team != 'Chelsea']
```

```python
getgoal = goals[goals.team == 'Chelsea']
bescored = goals[goals.team != 'Chelsea']
t = getgoal.minute[5]
t
```

```python
t = t.replace('[','')
t = t.replace(']','')
sample = t.split(', ')
sample
```

```python
new = []
for i in sample:
    new.append(int(i))
new
```

```python
def strtoint(group):
    group = group.replace('[','')
    group = group.replace(']','')
    ligroup = group.split(', ')
    intgroup = []
    for i in ligroup:
        if i != '':
            intgroup.append(int(i))
    return intgroup
getgoal_minute = getgoal.minute.map(strtoint)
getgoal_minute
```

```python
minute_any = [0,0,0,0,0,0]
for game in getgoal_minute:
    for time in game:
        if time <= 15:
            minute_any[0] += 1
        elif time <= 30:
            minute_any[1] += 1
        elif time <= 45:
            minute_any[2] += 1
        elif time <= 60:
            minute_any[3] += 1
        elif time <= 75:
            minute_any[4] += 1
        elif time < 91:
            minute_any[5] += 1
minute_any
```

```python
minute_series = Series(minute_any, index = ['0-15', '15-30', '30-45', '45-60', '60-75', '75-90'])
minute_series
```

```python
minute_series.sum()
```

```python
minute_rate = round(minute_series / minute_series.sum() * 100,1)
minute_rate
```

```python
map = minute_rate.plot(kind='bar')
map
```

```python
labels = minute_series.index
means = minute_series.values

x = np.arange(len(labels))
width = 0.5

fig, ax = plt.subplots()

rect = ax.bar(x, means, width)
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.set_title('Scoring time of 51 goals')
ax.legend()

def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                   xy=(rect.get_x() + rect.get_width() / 2, height),
                   xytext=(0, 3),
                   textcoords="offset points",
                   ha='center', va='bottom')
autolabel(rect)      

plt.show()
fig.savefig("getscore.png")
```
